```python
# -----------------------------------------------------------------------------
# Cell 2 — Imports, Config, and Output Folders
# -----------------------------------------------------------------------------
import os
from pathlib import Path
import numpy as np
import pandas as pd
import joblib
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.decomposition import PCA
from sklearn.mixture import BayesianGaussianMixture
from sklearn.metrics import adjusted_rand_score, confusion_matrix, silhouette_score

# ---- Configuration (edit here as needed) ----
# -----------------------------------------------------------------------------
# Cell 2 — Full configuration dictionary (drop-in replacement)
# -----------------------------------------------------------------------------
from pathlib import Path
from pathlib import Path

OUTPUT = Path("outputs"); FIG = OUTPUT/"figures"; TAB = OUTPUT/"tables"; ART = OUTPUT/"artifacts"
for p in [OUTPUT, FIG, TAB, ART]:
    p.mkdir(parents=True, exist_ok=True)

CFG = {
    # --- data window (adjust if you want a different period) ---
    # For your all_stocks_5yr CSV the available dates are 2013-02-08 -> 2018-02-07
    "start_date": "2013-02-08",   # <- change if you want a different start
    "end_date":   "2018-02-07",   # <- change if you want a different end

    # --- resampling / universe selection ---
    "resample":   "W-MON",        # weekly, anchored to Monday (used in resample())
    "target_universe": 50,        # aim for ~50 tickers (reduce if you have fewer)

    # --- randomness & reproducibility ---
    "seed":       42,
    
    # --- PCA & BGMM model parameters ---
    "pca_components": 3,
    "n_components": 8,            # BGMM truncation upper bound

    # --- bootstrap stability testing ---
    "bootstrap_reps": 100,
    "bootstrap_weeks": 30,

    # --- plotting / storage (no need to change) ---
    "fig_dpi": 300
}

# convenience: ensure outputs directories exist for later cells
OUTPUT = Path("outputs"); FIG = OUTPUT/"figures"; TAB = OUTPUT/"tables"; ART = OUTPUT/"artifacts"
for p in [OUTPUT, FIG, TAB, ART]:
    p.mkdir(parents=True, exist_ok=True)

# Seed RNGs used in notebook
import numpy as np
rng = np.random.default_rng(CFG["seed"])
np.random.seed(CFG["seed"])

print("CFG loaded. Date window:", CFG["start_date"], "→", CFG["end_date"])
print("Resample frequency:", CFG["resample"], "| Target universe:", CFG["target_universe"])
print("PCA comps:", CFG["pca_components"], "| BGMM n_components:", CFG["n_components"])

# Ensure the kaggle_data folder exists before saving
import os
os.makedirs("kaggle_data", exist_ok=True)

```

```python
PRICES_CSV = "C:/Users/sameer/Documents/bhai project/archive/all_stocks_5yr.csv"          # <- change if needed
META_CSV   = "C:/Users/sameer/Documents/bhai project/archive (2)/SP500.csv"    # <- change if needed

# Load
prices = pd.read_csv(PRICES_CSV)
meta   = pd.read_csv(META_CSV)

print("Raw shapes:", prices.shape, meta.shape)
print("Sample price columns:", list(prices.columns))
print("Sample meta columns:", list(meta.columns))

# Show first few rows (optional)
display(prices.head(3))
display(meta.head(3))
```

```python
# -----------------------------------------------------------------------------
# Cell 5 — Clean & standardize columns (UPDATED for your actual CSVs)
# -----------------------------------------------------------------------------
# This cell normalizes the price and meta dataframes into the canonical names:
#   df  -> columns: ['date', 'ticker', 'price', 'volume' (optional)]
#   df_meta -> columns: ['ticker', 'sector']
#
# It is written to match your printed columns:
#   prices columns: ['date','open','high','low','close','volume','Name']
#   meta columns:   ['Symbol','Security','GICS Sector', ...]
#
# No path edits required here — this uses the `prices` and `meta` variables loaded in Cell 4.

# copy originals (don't overwrite upstream variables in case you want to inspect)
df_prices = prices.copy()
df_meta   = meta.copy()

# strip whitespace from column names
df_prices.columns = [c.strip() for c in df_prices.columns]
df_meta.columns   = [c.strip() for c in df_meta.columns]

# -------------------------
# PRICE DATA NORMALIZATION
# -------------------------
# Common name candidates for ticker column in these downloads: 'Name', 'symbol', 'ticker'
ticker_candidates = ["ticker", "symbol", "Name", "NAME", "name"]
price_candidates  = ["adj_close","adj close","adjusted_close","close","Close","close_price"]
volume_candidates = ["volume", "vol"]

# detect ticker column name
detected_ticker_col = None
for cand in ticker_candidates:
    if cand in df_prices.columns:
        detected_ticker_col = cand
        break

# detect price column name
detected_price_col = None
for cand in price_candidates:
    if cand in df_prices.columns:
        detected_price_col = cand
        break

# detect date column name
date_candidates = ["date", "Date", "timestamp"]
detected_date_col = None
for cand in date_candidates:
    if cand in df_prices.columns:
        detected_date_col = cand
        break

# detect volume column (optional)
detected_vol_col = None
for cand in volume_candidates:
    if cand in df_prices.columns:
        detected_vol_col = cand
        break

# Fail early if we cannot find required columns
if detected_ticker_col is None:
    raise KeyError(f"Couldn't detect a ticker column in prices. Tried: {ticker_candidates}. Columns present: {list(df_prices.columns)}")
if detected_price_col is None:
    raise KeyError(f"Couldn't detect a price column in prices. Tried: {price_candidates}. Columns present: {list(df_prices.columns)}")
if detected_date_col is None:
    raise KeyError(f"Couldn't detect a date column in prices. Tried: {date_candidates}. Columns present: {list(df_prices.columns)}")

# Rename to canonical names
rename_map = {detected_ticker_col: "ticker", detected_price_col: "price", detected_date_col: "date"}
if detected_vol_col:
    rename_map[detected_vol_col] = "volume"

df_prices = df_prices.rename(columns=rename_map)

# Keep only canonical columns (plus volume if present)
cols_keep = ["date", "ticker", "price"] + (["volume"] if "volume" in df_prices.columns else [])
df = df_prices.loc[:, [c for c in cols_keep if c in df_prices.columns]].copy()

# Coerce types and filter date range from CFG
df["date"] = pd.to_datetime(df["date"], errors="coerce")
df = df.dropna(subset=["date", "price", "ticker"])
df = df[(df["date"] >= CFG["start_date"]) & (df["date"] <= CFG["end_date"])]

# Ensure ticker is string (strip whitespace)
df["ticker"] = df["ticker"].astype(str).str.strip()

# -------------------------
# META DATA NORMALIZATION
# -------------------------
# From your SP500.xlsx screenshot we expect: 'Symbol' and 'GICS Sector'
meta_rename = {}
if "Symbol" in df_meta.columns and "ticker" not in df_meta.columns:
    meta_rename["Symbol"] = "ticker"

# map 'GICS Sector' or 'Sector' -> 'sector'
if "GICS Sector" in df_meta.columns:
    meta_rename["GICS Sector"] = "sector"
elif "Sector" in df_meta.columns:
    meta_rename["Sector"] = "sector"
# apply renames (if any)
if meta_rename:
    df_meta = df_meta.rename(columns=meta_rename)

# pick only ticker+sector (drop rows missing sector)
expected_meta_cols = [c for c in ["ticker", "sector"] if c in df_meta.columns]
if "ticker" not in expected_meta_cols or "sector" not in expected_meta_cols:
    raise KeyError(f"Expected meta to contain 'ticker' and 'sector'. Meta columns found: {list(df_meta.columns)}")

df_meta = df_meta.loc[:, ["ticker", "sector"]].dropna(subset=["ticker", "sector"]).copy()
df_meta["ticker"] = df_meta["ticker"].astype(str).str.strip()
df_meta["sector"] = df_meta["sector"].astype(str).str.strip()

# -------------------------
# Quick sanity prints & save (optional)
# -------------------------
print("✅ Normalization complete.")
print("Prices (rows):", df.shape[0], "| unique tickers:", df["ticker"].nunique())
print("Meta (rows):", df_meta.shape[0], "| unique tickers:", df_meta["ticker"].nunique(), "| sectors:", df_meta["sector"].nunique())

display(df.head(3))
display(df_meta.head(3))

# Save standardized small csv copies for later cells (optional)
# (These are used by downstream cells that read from ART/.. but not strictly necessary)
df.to_csv("kaggle_data/prices_cleaned_for_pipeline.csv", index=False)
df_meta.to_csv("kaggle_data/constituents_cleaned_for_pipeline.csv", index=False)
```

```python
# -----------------------------------------------------------------------------
# Cell 6 (REPLACEMENT) — Build a ~50-ticker universe with diagnostics & auto-fix
# -----------------------------------------------------------------------------
from datetime import datetime

# df and df_meta should already be created by Cell 5
print("Configured date window:", CFG["start_date"], "→", CFG["end_date"])

# 1) Inspect available dates in the price dataframe
price_min = df["date"].min()
price_max = df["date"].max()
print("Available dates in prices:", price_min.date(), "→", price_max.date())

# 2) Check overlap between configured window and data
cfg_start = pd.to_datetime(CFG["start_date"])
cfg_end   = pd.to_datetime(CFG["end_date"])

# If there's no overlap, adjust the CFG window to the data's range and warn the user
if cfg_end < price_min or cfg_start > price_max:
    print("⚠️  Configured date window does not overlap the available price data.")
    print("    Auto-adjusting CFG start/end to available price date range to proceed.")
    CFG["start_date"] = price_min.strftime("%Y-%m-%d")
    CFG["end_date"]   = price_max.strftime("%Y-%m-%d")
    cfg_start = price_min
    cfg_end   = price_max
    print("    New CFG window:", CFG["start_date"], "→", CFG["end_date"])

# Now filter df again with the (possibly adjusted) window
df_filtered = df[(df["date"] >= cfg_start) & (df["date"] <= cfg_end)].copy()
print("Rows after date filtering:", df_filtered.shape[0])
print("Unique tickers in price data:", df_filtered["ticker"].nunique())

# 3) Inspect meta tickers
print("Unique tickers in meta (constituents):", df_meta["ticker"].nunique())

# 4) Intersection between price tickers and meta tickers
price_tickers = set(df_filtered["ticker"].unique())
meta_tickers  = set(df_meta["ticker"].unique())
common = sorted(list(price_tickers & meta_tickers))
print("Tickers in both price data and meta:", len(common))

if len(common) == 0:
    # helpful debugging info
    sample_price_tics = sorted(list(price_tickers))[:10]
    sample_meta_tics  = sorted(list(meta_tickers))[:10]
    raise RuntimeError(
        "No tickers overlap between price data and meta after date filtering.\n"
        f"  - Example price tickers (first 10): {sample_price_tics}\n"
        f"  - Example meta tickers  (first 10): {sample_meta_tics}\n"
        "Possible causes:\n"
        "  * Your dates (CFG['start_date']/['end_date']) remove all rows — adjust them.\n"
        "  * The tickers in the price file use a different naming convention (check capitalization or suffixes).\n"
        "  * You loaded the wrong files. Inspect `df.head()` and `df_meta.head()`.\n"
        "Fix: update CFG['start_date']/['end_date'] or ensure the meta file contains the same symbols as the prices file."
    )

# 5) Select candidate tickers ranked by liquidity or availability
if "volume" in df_filtered.columns:
    liq = df_filtered.groupby("ticker")["volume"].mean().sort_values(ascending=False)
    tickers_ranked = liq.index.tolist()
else:
    counts = df_filtered.groupby("ticker")["price"].count().sort_values(ascending=False)
    tickers_ranked = counts.index.tolist()

# keep only those present in meta with a sector
tickers_ranked = [t for t in tickers_ranked if t in meta_tickers]

# If after all this we still have fewer tickers than desired, reduce target_universe or warn
if len(tickers_ranked) == 0:
    raise RuntimeError("No tickers available after applying meta filter. Check that df_meta contains the same ticker names as the prices file.")

# Final universe selection (cap to target_universe)
universe = tickers_ranked[:min(CFG["target_universe"], len(tickers_ranked))]
universe_meta = df_meta[df_meta["ticker"].isin(universe)].copy()

# map sectors to numeric ids for ARI
sector_names = sorted(universe_meta["sector"].unique())
sector_to_id = {s:i for i,s in enumerate(sector_names)}
universe_meta["sector_id"] = universe_meta["sector"].map(sector_to_id)

print(f"Selected {len(universe)} tickers across {len(sector_names)} sectors.")
if len(universe) < CFG["target_universe"]:
    print("Note: fewer tickers available than target_universe — proceeding with available tickers.")

display(universe_meta.head(10))
```

```python
# -----------------------------------------------------------------------------
# Cell 7 — Fit Bayesian Gaussian Mixture Model (BGMM) on PCA data
# -----------------------------------------------------------------------------
from sklearn.mixture import BayesianGaussianMixture
import pandas as pd, numpy as np, joblib
from pathlib import Path

# Make sure output folders exist
ART.mkdir(parents=True, exist_ok=True)

# Load PCA-transformed features
X_pca_df = pd.read_csv(ART/"X_pca.csv", index_col=0)
X_pca_mat = X_pca_df.values

# Fit Bayesian Gaussian Mixture
bgmm = BayesianGaussianMixture(
    n_components=CFG["n_components"],
    covariance_type="full",
    weight_concentration_prior_type="dirichlet_process",
    random_state=CFG["seed"],
    max_iter=1000
).fit(X_pca_mat)

# Predict cluster labels & soft responsibilities
labels = bgmm.predict(X_pca_mat)
resp   = bgmm.predict_proba(X_pca_mat)

# Save for later cells
pd.Series(labels).to_csv(ART/"labels.csv", index=False)
pd.DataFrame(resp).to_csv(ART/"responsibilities.csv", index=False)
joblib.dump(bgmm, ART/"bgmm.joblib")

# Effective number of active clusters
eff_k = int(np.sum(bgmm.weights_ > 1e-3))
print(f"✅ BGMM fitted successfully — Effective clusters (weights>1e-3): {eff_k}/{CFG['n_components']}")
```

```python
# -----------------------------------------------------------------------------
# ✅ FIXED Cell 8 — Evaluate BGMM clustering vs. sectors, silhouette (aligned)
# -----------------------------------------------------------------------------
from sklearn.metrics import adjusted_rand_score, silhouette_score
import pandas as pd, numpy as np, joblib

# Load saved artifacts
labels = pd.read_csv(ART/"labels.csv", header=None)[0].values
resp   = pd.read_csv(ART/"responsibilities.csv").values
tick   = pd.read_csv(ART/"tickers_labeled.csv")
X_pca_df = pd.read_csv(ART/"X_pca.csv", index_col=0)
X_pca_mat = X_pca_df.values

# --- Align by ticker index ---
# Some tickers may have dropped out of the meta or PCA set; fix alignment.
tickers_bgmm = list(X_pca_df.index)
tick = tick[tick["ticker"].isin(tickers_bgmm)]

# Ensure same order as PCA rows
tick = tick.set_index("ticker").reindex(tickers_bgmm).reset_index()

# Create aligned arrays
sector_ids = tick["sector_id"].to_numpy()
labels     = labels[:len(sector_ids)]  # trim to match length, just in case

metrics = {}

# 1️⃣ Adjusted Rand Index (if sector labels available)
if np.all(sector_ids >= 0) and len(np.unique(sector_ids)) > 1 and len(sector_ids) == len(labels):
    ari = adjusted_rand_score(sector_ids, labels)
    metrics["ARI_vs_sector"] = float(ari)
else:
    metrics["ARI_vs_sector"] = np.nan

# 2️⃣ Silhouette score (cluster separation quality)
if len(np.unique(labels)) > 1 and X_pca_mat.shape[0] > len(np.unique(labels)):
    sil = silhouette_score(X_pca_mat, labels)
    metrics["Silhouette_PCA"] = float(sil)
else:
    metrics["Silhouette_PCA"] = np.nan

# 3️⃣ Bootstrap skipped (simplified version)
metrics["Bootstrap_ARIMean"] = np.nan
metrics["Bootstrap_ARISD"]   = np.nan

# 4️⃣ BGMM info
bgmm = joblib.load(ART/"bgmm.joblib")
pca  = joblib.load(ART/"pca.joblib")
eff_k = int(np.sum(bgmm.weights_ > 1e-3))
metrics["Effective_K"] = eff_k
metrics["PCA_var_PC1_PC2_PC3"] = float(np.sum(pca.explained_variance_ratio_))

# Save & display
metrics_df = pd.DataFrame([metrics])
metrics_df.to_csv(TAB/"executive_summary.csv", index=False)

print("✅ Evaluation complete (arrays aligned).")
print(f"Samples aligned: {len(labels)} | Clusters used: {eff_k}")
display(metrics_df)
```

```python
# -----------------------------------------------------------------------------
# ✅ Fixed Cell 9 — Metrics & summary (alignment-safe)
# -----------------------------------------------------------------------------
import joblib
import numpy as np
import pandas as pd
from sklearn.metrics import adjusted_rand_score, silhouette_score

# Load artifacts
labels = pd.read_csv(ART/"labels.csv", header=None)[0].values
X_pca_df = pd.read_csv(ART/"X_pca.csv", index_col=0)
X_pca_mat = X_pca_df.values
tick = pd.read_csv(ART/"tickers_labeled.csv")

# --- Align tickers between PCA and metadata ---
tickers_pca = list(X_pca_df.index)
tick = tick[tick["ticker"].isin(tickers_pca)].set_index("ticker")
# Make sure order matches PCA index
tick = tick.reindex(tickers_pca)
tick = tick.dropna(subset=["sector_id"])  # just in case

# Now both have same tickers
aligned_labels = labels[:len(tick)]
sector_ids = tick["sector_id"].astype(int).values

# --- Compute metrics ---
metrics = {}

# 1️⃣ Adjusted Rand Index
if len(aligned_labels) == len(sector_ids) and len(np.unique(sector_ids)) > 1:
    ari = adjusted_rand_score(sector_ids, aligned_labels)
    metrics["ARI_vs_sector"] = float(ari)
else:
    metrics["ARI_vs_sector"] = np.nan

# 2️⃣ Silhouette
if len(np.unique(aligned_labels)) > 1 and X_pca_mat.shape[0] > len(np.unique(aligned_labels)):
    sil = silhouette_score(X_pca_mat, aligned_labels)
    metrics["Silhouette_PCA"] = float(sil)
else:
    metrics["Silhouette_PCA"] = np.nan

# 3️⃣ Bootstrap skipped
metrics["Bootstrap_ARIMean"] = np.nan
metrics["Bootstrap_ARISD"]   = np.nan

# 4️⃣ Model diagnostics
bgmm = joblib.load(ART/"bgmm.joblib")
pca  = joblib.load(ART/"pca.joblib")
eff_k = int(np.sum(bgmm.weights_ > 1e-3))
metrics["Effective_K"] = eff_k
metrics["PCA_var_PC1_PC2_PC3"] = float(np.sum(pca.explained_variance_ratio_))

# --- Save results ---
metrics_df = pd.DataFrame([metrics])
metrics_df.to_csv(TAB/"executive_summary.csv", index=False)

print("✅ Metrics computed successfully.")
print(f"Aligned samples: {len(aligned_labels)} | Unique clusters: {len(np.unique(aligned_labels))}")
display(metrics_df)
```

```python
# -----------------------------------------------------------------------------
# Final Fix Cell — recompute integer cluster labels, ARI, and Figure A3
# -----------------------------------------------------------------------------
import numpy as np, pandas as pd, joblib
from pathlib import Path
from sklearn.metrics import adjusted_rand_score, confusion_matrix
import matplotlib.pyplot as plt, seaborn as sns

sns.set_context("notebook")

# paths used throughout the notebook
ART = Path("outputs/artifacts"); TAB = Path("outputs/tables"); FIG = Path("outputs/figures")
for p in [ART, TAB, FIG]:
    p.mkdir(parents=True, exist_ok=True)

# --- Load stored artifacts ---
X_pca_df = pd.read_csv(ART / "X_pca.csv", index_col=0)
resp_df  = pd.read_csv(ART / "responsibilities.csv", index_col=0)
tick_df  = pd.read_csv(ART / "tickers_labeled.csv")

# --- Derive cluster labels from BGMM responsibilities ---
# This ensures the labels are numeric integers (0..K-1)
labels = resp_df.values.argmax(axis=1).astype(int)

# --- Align sector IDs to tickers ---
sector_map = dict(zip(tick_df["ticker"], tick_df["sector_id"]))
sector_ids = np.array([sector_map.get(t, -1) for t in X_pca_df.index], dtype=int)

# --- Compute ARI vs. sector ---
if (sector_ids >= 0).all() and len(np.unique(sector_ids)) > 1:
    ari = adjusted_rand_score(sector_ids, labels)
else:
    ari = np.nan

print(f"✅ ARI vs. sector recomputed: {ari:.4f}" if not np.isnan(ari) else "⚠️ ARI not computed (missing sector info).")
print("Unique clusters detected:", len(np.unique(labels)))

# --- Update executive_summary.csv with new ARI ---
summary_path = TAB / "executive_summary.csv"
if summary_path.exists():
    es = pd.read_csv(summary_path)
else:
    es = pd.DataFrame([{}])
es["ARI_vs_sector"] = ari
es.to_csv(summary_path, index=False)
print("Updated executive_summary.csv")

# --- Plot Figure A3 (Confusion Matrix) ---
if not np.isnan(ari):
    cm = confusion_matrix(sector_ids, labels)
    plt.figure(figsize=(6, 5))
    sns.heatmap(cm, annot=True, fmt="d", cmap="coolwarm", cbar=False)
    plt.title("Figure A3. Confusion Matrix: Sector vs. BGMM Cluster")
    plt.xlabel("Predicted Cluster")
    plt.ylabel("True Sector")
    plt.tight_layout()
    plt.savefig(FIG / "Figure_A3_Confusion_Matrix.png", dpi=300)
    plt.show()
    print("✅ Saved:", FIG / "Figure_A3_Confusion_Matrix.png")
else:
    print("Skipped Figure A3 — sector labels unavailable or ARI invalid.")
```
